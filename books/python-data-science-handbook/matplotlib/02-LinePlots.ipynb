# Line Plots

Now that we've got acquainted with Matplotlib, let's take a look at some simple line plots. We'll start by setting up the notebook for plotting:

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

For all Matplotlib plots, we start by creating a figure and an axes:

```python
fig = plt.figure()
ax = plt.axes()
```

In Matplotlib, the _figure_ (an instance of the class `plt.Figure`) can be thought of as a single container that contains all the objects representing axes, graphics, text, and labels. The _axes_ (an instance of the class `plt.Axes`) is a bounding box with ticks and labels (as seen above), which will eventually contain the plot elements.

With the axes created, we can use the `ax.plot` function to plot some data. Let's start with a sinusoid:

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

As seen in the previous notebook, we can also use the pylab interface and let the figure and axes be created automatically for us in the background:

```python
plt.plot(x, np.sin(x));
```

To create a single figure with multiple lines, we can simply call the `plot` function multiple times:

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

## Line Colors and Styles

The `plt.plot()` function takes additional arguments that can be used to specify colors and styles. To adjust the color, we can use the `color` keyword, which accepts the specification of colors in a variety of ways:

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

By default, Matplotlib will automatically cycle through a set of default colors for multiple lines in the same plot.

Similarly, the style can be adjusted using the `linestyle` keyword:

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

The `linestyle` and `color` codes can be combined into a single non-keyword argument to the `plt.plot()` function:

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
```

These single-character color codes reflect the standard abbreviations in the RGB (Red/Green/Blue) and CMYK (Cyan/Magenta/Yellow/blacK) color systems.

## Axes Limits

Matplotlib chooses by default axes limits for the plots, but we can have a finer control through the use of the `plt.xlim()` and `plt.ylim()` methods:

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

We can even display the axis in reverse by changing the order of the arguments:

```python
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

A useful related method is `plt.axis()`, which allows us to set the `x` and `y` limits with a single call, by passing a list that specifies `[xmin, xmax, ymin, ymax]`:

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

It also allows us to do things like automatically tighten the bound around the current plot:

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

We can even ensure an equal aspect ratio so that one unit in `x` is equal to one unit in `y`:

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

## Labeling Plots

Let's take a quick look at the labeling of plots: titles, axis labels, and simple legends.

Titles and axis labels can be quickly set through the appropriate methods:

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

We can adjust the position, size and style of these labels using optional arguments.

When multiple lines are being shown within a single axes, it can be useful to create a plot legend that labels each line type. We can specify the label of each line using the `label` keyword of the plot function. The legend is created via the `plt.legend()` method:

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend();
```

## Matplotlib Gotchas

While most `plt` functions translate directly to `ax` methods, this is not the case for all commands. For transitioning between MATLAB-style functions and object-oriented methods, we need to make the following changes:

- `plt.xlabel()` &rarr; `ax.set_xlabel()`
- `plt.ylabel()` &rarr; `ax.set_ylabel()`
- `plt.xlim()` &rarr; `ax.set_xlim()`
- `plt.ylim()` &rarr; `ax.set_ylim()`
- `plt.title()` &rarr; `ax.set_title()`

In the object-oriented interface, it is often more convenient to use the `ax.set()` method to set all these properties at once:

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```
